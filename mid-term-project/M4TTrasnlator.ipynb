```python
import gradio as gr

from gradio.components import dropdown
from gradio.components.audio import Audio


from gradio_client import Client

client = Client("https://facebook-seamless-m4t.hf.space/")

lang_list = ("Afrikaans",
     "Amharic",
     "Modern Standard Arabic",
     "Moroccan Arabic",
     "Egyptian Arabic",
     "Assamese",
     "Asturian",
     "North Azerbaijani",
     "Belarusian",
     "Bengali",
     "Bosnian",
     "Bulgarian",
     "Catalan",
     "Cebuano",
     "Czech",
     "Central Kurdish",
     "Mandarin Chinese",
     "Welsh",
     "Danish",
     "German",
     "Greek",
     "English",
     "Estonian",
     "Basque",
     "Finnish",
     "French",
     "West Central Oromo",
     "Irish",
     "Galician",
     "Gujarati",
     "Hebrew",
     "Hindi",
     "Croatian",
     "Hungarian",
     "Armenian",
     "Igbo",
     "Indonesian",
     "Icelandic",
     "Italian",
     "Javanese",
     "Japanese",
     "Kamba",
     "Kannada",
     "Georgian",
     "Kazakh",
     "Kabuverdianu",
     "Halh Mongolian",
     "Khmer",
     "Kyrgyz",
     "Korean",
     "Lao",
     "Lithuanian",
     "Luxembourgish",
     "Ganda",
     "Luo",
     "Standard Latvian",
     "Maithili",
     "Malayalam",
     "Marathi",
     "Macedonian",
     "Maltese",
     "Meitei",
     "Burmese",
     "Dutch",
     "Nepali",
     "Nyanja",
     "Occitan",
     "Odia",
     "Punjabi",
     "Southern Pashto",
     "Western Persian",
     "Polish",
     "Portuguese",
     "Romanian",
     "Russian",
     "Slovak",
     "Slovenian",
     "Shona",
     "Sindhi",
     "Somali",
     "Spanish",
     "Serbian",
     "Swedish",
     "Swahili",
     "Tamil",
     "Telugu",
     "Tajik",
     "Tagalog",
     "Thai",
     "Turkish",
     "Ukrainian",
     "Urdu",
     "Northern Uzbek",
     "Vietnamese",
     "Xhosa",
     "Yoruba",
     "Cantonese",
     "Colloquial Malay",
     "Standard Malay",
     "Zulu",
)

# Called when Text to Text & Speech is required
def T2ST(source_lang: str, target_lang: str, text_data: str):
    task_name = 'T2ST'
    audio_source = ''
    input_speech = None
    input_speech1 = None
    input_text = text_data
    source_lang = source_lang
    target_lang = target_lang 
    result = client.predict(
          task_name,
          audio_source,
          input_speech,
          input_speech1,
          input_text,
          source_lang,
          target_lang ,
          api_name="/run"
        )
    return result

# Called when Speech to Speech & Text is required
def S2ST(source_lang: str, target_lang: str, audio):
    task_name = 'T2ST'
    audio_source = ''
    input_speech = None
    input_speech1 = None
    input_text = ""
    source_lang = ""
    target_lang = ""
    result = client.predict(
          task_name,
          audio_source,
          input_speech,
          input_speech1,
          input_text,
          source_lang,
          target_lang ,
          api_name="/run"
        )
    return result

# Function with hardcoded values to display how Speech to Speech Translation Works
def S2STT(target_lang: str, audio):
    task_name = 'T2ST'
    audio_source = 'file'
    input_speech = None
    input_speech1 = 'https://huggingface.co/spaces/facebook/seamless_m4t/resolve/main/assets/sample_input_2.mp3'
    input_text = ''
    source_lang = 'English'
    target_lang = target_lang
    result = client.predict(
          task_name,
          audio_source,
          input_speech,
          input_speech1,
          input_text,
          source_lang,
          target_lang ,
          api_name="/run"
        )
    return result[0]

# Setting up inputs and outputs for Text to Text & Speech Translation
inputs = [ gr.Dropdown(label="Source Language", choices=(lang_list), value="Select Source Language"),
           gr.Dropdown(label="Target Language", choices=(lang_list), value="Select Target Language"),
           gr.Text(label="Input", max_lines=3)
           ]
outputs = gr.Audio(label="Translated Audio")

# TEXT TO TEXT & TEXT TO SPEECH TRANSLATION INTERFACE
t2stt_interface = gr.Interface(fn=T2ST,
        inputs=inputs,
        outputs=["audio","text"],
        title="""Text to Text & Text to Speech 
                Translation Using Seamless M4t Facebook Translation""",
        description="Select Source Language, Target Language and Text to Translate",
        allow_flagging="never",
    )


# Setting up inputs and outputs for Speech to Speech & Text Translation
inputs_s2s = [ 
    #gr.Dropdown(label="Source Language", choices=(lang_list), value="Select Source Language"),
           gr.Dropdown(label="Target Language", choices=(lang_list), value="Select Target Language"),
           gr.Microphone(label="Input")
           ]
outputs_s2s = gr.Audio(label="Output")

# SPEECH TO TEXT & SPEECH TO SPEECH TRANSLATION INTERFACE
s2stt_interface = gr.Interface(fn=S2STT,
      inputs=inputs_s2s,
      outputs=["audio"],
      title="""Speech to Speech & Speech
                Translation Using Seamless M4t Facebook Translation. Currently I have put a hardcoded sound 
                file link so I can give you a demo how it will work. I am working to push the value recorded from the
                Microphone to be Translated.""",
        description="Select Source Language, Target Language and Text to Translate",
        allow_flagging="never",
)

# Setting up Tab Interface to launch.
applaunch = gr.TabbedInterface([t2stt_interface, s2stt_interface], ["Text to Text & Text to Speech Translation", "Speech to Speech & Text Translation"])

applaunch.queue(max_size=50).launch(share=True)
```
